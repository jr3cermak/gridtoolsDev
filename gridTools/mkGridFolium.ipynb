```python
import folium
from folium.plugins import Draw

m = folium.Map(location=[58.5, -149.0], zoom_start=6)

draw = Draw(export=True)

draw.add_to(m)

m
```

```python
print(m)
print(dir(m))
print(m.to_json())
```

```python
# Try Mercator (epsg:3857)
m = folium.Map(location=[58.5, -149.0], zoom_start=6, crs="EPSG3857")

draw = Draw(export=True)

draw.add_to(m)

m
```

```python
# Try Lambert Conformal Conic (lcc)
# +proj=lcc +lon_0=-90 +lat_1=33 +lat_2=45

m = folium.Map(location=[58.5, -149.0], zoom_start=6, crs="+proj=lcc +lon_0=-90 +lat_1=33 +lat_2=45")

draw = Draw(export=True)

draw.add_to(m)

m
```

```python
# Try Polar Stereographic
# +proj=stere +lat_0=90 +lat_ts=75

m = folium.Map(location=[58.5, -149.0], zoom_start=6, crs="+proj=stere +lat_0=90 +lat_ts=75")

draw = Draw(export=True)

draw.add_to(m)

m
```

```python
#conda: legacyTools gridTools

from owslib.wms import WebMapService
```

```python
wms = WebMapService('https://tigerweb.geo.census.gov/arcgis/services/TIGERweb/tigerWMS_Current/MapServer/WMSServer',
                    version='1.1.1')
print(wms.identification.type)
print(wms.identification.version)
print(wms.identification.title)
print(wms.identification.abstract)
```

```python
#list(wms.contents)
```

```python
print(wms['States'].crsOptions)
print(wms['States'].styles)
print(wms['States'].boundingBoxWGS84)
```

```python
[op.name for op in wms.operations]
```

```python
wms.getOperationByName('GetMap').methods
```

```python
wms.getOperationByName('GetMap').formatOptions
```

```python
img = wms.getmap(layers=['States'],
                 styles=['default'],
                 srs='EPSG:4326',
                 bbox=(-175, 50, -50, 80),
                 size=(600, 300),
                 format='image/jpeg',
                 transparent=True
)
out = open('state_test.jpg', 'wb')
out.write(img.read())
out.close()

from IPython.display import Image
Image(filename = 'state_test.jpg', width = 600, height = 300)
```

```python
# Grid Generator Map Server
wms = WebMapService('https://recon.lccllc.info/ggms4326',
                    version='1.1.1')
print(wms.identification.type)
print(wms.identification.version)
print(wms.identification.title)
print(wms.identification.abstract)
img = wms.getmap(layers=['earth,grid'],
                 styles=['default'],
                 srs='EPSG:4326',
                 bbox=(-180, 50, -100, 80),
                 size=(600, 300),
                 format='image/jpeg',
                 transparent=True
)
out = open('grid_test.jpg', 'wb')
out.write(img.read())
out.close()

from IPython.display import Image
Image(filename = 'grid_test.jpg', width = 600, height = 300)
```

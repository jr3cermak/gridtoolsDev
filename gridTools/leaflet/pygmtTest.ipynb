```python
# conda: legacyTools
# Install of GMT 6+ is required
import pygmt
```

```python
# Overall GMT options
# width = 12c = 12 centimeters
# region = [-130, -70, 24, 52] = xmin, xmax, ymin, ymax = EW/NS
#   May be bottom-left and top-right corners using "10/35/20/45+r"
#   Grid generation: -Rcodex0/y0/nx/ny
#     code: two letters
#       LCR [left, center, right]
#       TMB [top, middle, bottom]
```

```python
# Mercator cylindrical
# pygmt: m[lon0[/lat0]]/scale or M[lon0[/lat0]/]width
```

```python
# California (Niki)
# dLon = 20.0
# dLat = 30.0
# Parameters
# lon0 = 230.0
# lat0 = 40.0
# Angle
# tilt = 30.0
# -Sj ?
```

```python
lon_span=20.
lat_span=30.

lon0=230.0
lat0=40.0
tilt=30.0

fig = pygmt.Figure()
fig.coast(
    shorelines="1/0.5p",
    region="CM%d/%d/%d/%d -I1/1" % (lon0, lat0, lon_span, lat_span),
    projection="M12c",
    land="gray",
    borders=["1/thick,black", "2/thin,black"],
    frame="afg",
)

fig.show()
```
